First of all, we create an OrderBookClient. Here we use BTCUSDT as our example

```python
from order_book_client_dy import OrderBookClient
import utils
symbol = 'BTCUSDT'
OBClient_0 = OrderBookClient([symbol], utils.process_message)
```

This is a snapshot of orderbook requested from Binance

```python
data = OBClient_0.get_snapshot(symbol)
data
```

Then, we use funcions in dicionarize to make the message above into a dictionary, which is much easier for further updating

```python
import dictionarize
orderbook = dictionarize.message_to_orderbookd(data)
orderbook
```

Update local orderbook example, here we assume the sell order at price 58737.99 is cancelled:

```python
diff = {'a': [['58737.99','0']], 'b':[]}
dictionarize.update_orderbook(orderbook, diff)
```
